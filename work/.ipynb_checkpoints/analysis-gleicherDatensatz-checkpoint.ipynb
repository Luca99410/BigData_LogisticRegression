```python
# [1] Imports, Utility-Funktionen & Pfade
import os, time, uuid, math, datetime as dt
from pathlib import Path
from typing import Dict, Any, Optional, List, Tuple, Iterable

import pandas as pd
import numpy as np

from pyspark.sql import SparkSession, DataFrame
from pyspark.sql import functions as F
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator

import matplotlib.pyplot as plt

# Pfade
ROOT = Path("/home/jovyan/work").resolve()
DATA_DIR = ROOT / "data"
RESULTS_DIR = ROOT / "results"
DATA_DIR.mkdir(parents=True, exist_ok=True)
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

# Zeitstempel & Ordner
def now_ts() -> str:
    return dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

def ts_slug() -> str:
    return dt.datetime.now().strftime("%Y%m%d_%H%M%S")

def make_run_dir(label:str, base:Path=RESULTS_DIR) -> Path:
    folder = base / f"{ts_slug()}_{label}"
    folder.mkdir(parents=True, exist_ok=True)
    return folder

# CSV persistieren
def save_series_csv(rows: List[Dict[str, Any]], path: Path) -> str:
    df = pd.DataFrame(rows)
    path.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(path, index=False)
    return str(path)

# „schöne“ Zahl
def human(x):
    if isinstance(x, float):
        return float(f"{x:.4g}")
    return x

print("Dirs:", DATA_DIR, RESULTS_DIR)
```

```python
# [2] SparkSession-Helfer
from pyspark.sql import SparkSession

def build_spark(
    mode: str = "cluster",
    app_name: str = "LogReg-Benchmark",
    driver_mem: str = "2g",
    executor_mem: str = "3g",
    executor_cores: int = 4,   # 4 bei 1 Worker, 2 bei 2 Workern
    cores_max: int = 4,        # total cores = 4
    ui_base_port: int = 4040,
    shuffle_parts: int = 8
) -> SparkSession:
    """
    Standalone-Cluster (spark://spark-master:7077). In Standalone startet genau EIN Executor pro Worker.
    Für 1 Worker sollte executor_cores = 4 gesetzt werden; für 2 Worker executor_cores = 2.
    """
    master = "local[*]" if mode == "local" else "spark://spark-master:7077"
    spark = (
        SparkSession.builder
        .appName(app_name)
        .master(master)
        .config("spark.driver.memory", driver_mem)
        .config("spark.executor.memory", executor_mem)
        .config("spark.executor.cores", executor_cores)
        .config("spark.cores.max", cores_max)
        .config("spark.sql.shuffle.partitions", shuffle_parts)
        .config("spark.driver.host", os.environ.get("SPARK_DRIVER_HOST", "spark-jupyter"))
        .config("spark.driver.bindAddress", os.environ.get("SPARK_DRIVER_BIND_ADDRESS", "0.0.0.0"))
        .config("spark.ui.port", ui_base_port)
        .config("spark.ui.prometheus.enabled", "false")
        .config("spark.local.dir", "/home/jovyan/work/tmp")
        .getOrCreate()
    )
    print("Spark version:", spark.version)
    print("Master:", master)
    print("Exec mem:", executor_mem, "Exec cores:", executor_cores, "Cores max:", cores_max)
    return spark

# Beispiel-Init (für 1 Worker: executor_cores=4; für 2 Worker: 2)
spark = build_spark(
    mode="cluster",
    app_name="LogReg-Benchmark",
    driver_mem="2g",
    executor_mem="3g",
    executor_cores=4,  # <- bei 2 Workern hier 2 setzen
    cores_max=4,
    ui_base_port=4040,
    shuffle_parts=8
)

# [2b] Mini-Sanity
rdd = spark.sparkContext.parallelize(range(10), 4)
print("Sum =", rdd.sum())

# [2c] Cluster-Health-Check (mit korrekter MB->GiB Umrechnung im Master-JSON)
import json, urllib.request
sc = spark.sparkContext
print("appId:", sc.applicationId)
print("master:", sc.master)
print("ui:    ", sc.uiWebUrl)
print("defaultParallelism:", sc.defaultParallelism)

# Standalone Master JSON
def master_overview():
    for host in ("spark-master", "localhost"):
        try:
            with urllib.request.urlopen(f"http://{host}:8080/json", timeout=2) as r:
                data = json.loads(r.read().decode("utf-8"))
                workers = data.get("workers", [])
                tot_cores = sum(w.get("cores", 0) for w in workers)
                tot_mem_gib = sum(w.get("memory", 0) for w in workers)/1024  # MB -> GiB
                print(f"\nMaster {host}: {len(workers)} Worker, Cores={tot_cores}, Mem≈{tot_mem_gib:.2f} GiB")
                for w in workers:
                    print(f"  {w.get('id')}  cores={w.get('cores')}  mem={w.get('memory')/1024:.2f} GiB  {w.get('host')}")
                return
        except Exception:
            pass
    print("\nMaster-JSON nicht erreichbar (Port 8080?)")
master_overview()
```

```python
# [3] Deterministische binäre Datensätze – robustes Parquet-Handling für verteiltes Schreiben

from pyspark.sql import functions as F
from pyspark.ml.feature import VectorAssembler
import numpy as np
import shutil

def dataset_path(N:int, d:int, seed:int=42) -> Path:
    # Parquet als ORDNER; gleiche Pfade in allen Containern gemountet!
    return DATA_DIR / f"binomial_N{int(N)}_d{int(d)}_seed{int(seed)}.parquet"

def _rm_local_path(p: Path):
    if not p.exists():
        return
    if p.is_file() or p.is_symlink():
        p.unlink(missing_ok=True)
    else:
        shutil.rmtree(p, ignore_errors=True)

def _hadoop_fs(spark):
    return spark._jvm.org.apache.hadoop.fs.FileSystem.get(spark._jsc.hadoopConfiguration())

def _h_path(spark, p: Path):
    # explizit als file:-URI
    return spark._jvm.org.apache.hadoop.fs.Path("file:" + str(p))

def _h_delete(spark, p: Path):
    fs = _hadoop_fs(spark)
    jdst = _h_path(spark, p)
    if fs.exists(jdst):
        fs.delete(jdst, True)

def _parquet_exists_nonempty(spark, p: Path) -> bool:
    fs = _hadoop_fs(spark)
    jdst = _h_path(spark, p)
    if not fs.exists(jdst):
        return False
    # Prüfen, ob Verzeichnis und etwas drin (SUCCESS oder part-*)
    if fs.isFile(jdst):
        return False
    it = fs.listStatus(jdst)
    # irgendwas drin? (SUCCESS/part-*.parquet)
    return len(it) > 0

def make_or_load_dataset(spark, N:int, d:int, seed:int=42, repartition:int=0) -> DataFrame:
    """
    Erzeugt deterministisch X∈R^{N×d}, y∈{0,1}^N und persistiert als Parquet-Ordner
    unter DATA_DIR. Pfad muss von allen Executorn erreichbar sein (Volumes in docker-compose).
    """
    p = dataset_path(N, d, seed)
    DATA_DIR.mkdir(parents=True, exist_ok=True)

    # Read-Path (falls bereits gültig)
    if _parquet_exists_nonempty(spark, p):
        df = spark.read.parquet(str(p))
        if repartition and repartition > 0:
            df = df.repartition(int(repartition))
        return df

    # Falls lokale Datei oder defekter Ordner existiert: lokal + Hadoop-View löschen
    if p.exists():
        _rm_local_path(p)
    _h_delete(spark, p)

    # X erzeugen
    base = spark.range(0, int(N)).withColumnRenamed("id", "row_id")
    cols = []
    for j in range(int(d)):
        base = base.withColumn(f"x{j}", F.randn(seed + j))
        cols.append(f"x{j}")

    rng = np.random.RandomState(seed)
    w = rng.normal(0.0, 1.0, size=(int(d),))
    b = float(rng.normal(0.0, 0.5, size=()))

    z = None
    for j in range(int(d)):
        term = F.col(f"x{j}") * float(w[j])
        z = term if z is None else (z + term)
    z = z + F.lit(b)

    Xy = base.withColumn("p", (1.0/(1.0 + F.exp(-z)))) \
             .withColumn("u", F.rand(seed + 10_000))

    assembler = VectorAssembler(inputCols=cols, outputCol="features")
    df = assembler.transform(Xy) \
                  .withColumn("label", F.when(F.col("u") < F.col("p"), F.lit(1.0)).otherwise(F.lit(0.0))) \
                  .select("label", "features")

    # In TEMP-Ziel schreiben (vermeidet Kollisionen), dann atomar umbenennen
    tmp_p = Path(str(p) + ".__tmp__")
    if tmp_p.exists():
        _rm_local_path(tmp_p)
    _h_delete(spark, tmp_p)

    df.write.mode("overwrite").parquet(str(tmp_p))

    fs = _hadoop_fs(spark)
    jtmp = _h_path(spark, tmp_p)
    jdst = _h_path(spark, p)
    _h_delete(spark, p)  # sicherheitshalber
    if not fs.rename(jtmp, jdst):
        # Fallback: lokal löschen und neu schreiben (sehr selten)
        _rm_local_path(p)
        df.write.mode("overwrite").parquet(str(p))

    # Validierung
    df = spark.read.parquet(str(p))
    if repartition and repartition > 0:
        df = df.repartition(int(repartition))
    return df

print("OK: robustes Dataset-Handling aktiv (gemeinsamer Mount + Temp-Rename).")
```

```python
# [3b] Cleanup (optional) – entfernt nur defekte/leerstehende Parquet-Ordner
# nutzt die in [3] definierten Helfer: _parquet_exists_nonempty, _rm_local_path, _h_delete

def cleanup_datasets(spark, pattern: str = "binomial_N*_d*_seed*.parquet", remove_all: bool = False, verbose: bool = True):
    """
    Entfernt defekte/leerstehende Parquet-Ziele unter DATA_DIR.
    - remove_all=False: löscht nur Einträge, die (a) als Datei existieren ODER (b) leere Parquet-Ordner sind.
    - remove_all=True: löscht ALLE Treffer des Patterns (Hard Reset).
    """
    removed, kept = 0, 0
    for item in DATA_DIR.glob(pattern):
        try:
            if remove_all:
                if verbose: print(f"Hard-Reset: entferne {item}")
                _rm_local_path(item)
                _h_delete(spark, item)
                removed += 1
                continue

            # Soft-Cleanup: nur offensichtliche Problemfälle entfernen
            is_bad = False
            if item.is_file():
                is_bad = True
            else:
                # Ordner: existiert er aus Sicht von Hadoop-FS und hat er Inhalte?
                if not _parquet_exists_nonempty(spark, item):
                    is_bad = True

            if is_bad:
                if verbose: print(f"Cleanup: entferne defekten Eintrag {item}")
                _rm_local_path(item)
                _h_delete(spark, item)
                removed += 1
            else:
                kept += 1
        except Exception as e:
            print(f"Warnung: Konnte {item} nicht bereinigen: {e}")

    print(f"Cleanup: removed={removed}, kept={kept}, pattern='{pattern}', remove_all={remove_all}")

# Optional aufrufen:
# Nur defekte Einträge entfernen:
# cleanup_datasets(spark)
# Oder alles (Hard Reset):
# cleanup_datasets(spark, remove_all=True)
```

```python
# [4] Hyperparameter-Preset (binäre logistische Regression)
# L2-Regularisierung (praxisnah), Solver: L-BFGS

LR_PRESET = dict(
    regParam=0.01,          # L2-Stärke
    elasticNetParam=0.0,    # 0 = reines L2
    maxIter=100,
    tol=1e-6,
    standardization=True,
    family="binomial",
    aggregationDepth=2,
    fitIntercept=True
)

evaluator_bin = BinaryClassificationEvaluator(
    labelCol="label",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderROC"
)
print("LR_PRESET:", LR_PRESET)
```

```python
# [5] Einzelexperiment: fester Datensatz (geladen aus Parquet) -> LR-Fit + Kennzahlen

def run_logreg_experiment(
    spark,
    N:int,
    d:int,
    *,
    seed_data:int = 42,                 # identischer Datensatz über Runs
    test_fraction: Optional[float] = None,  # z.B. 0.2 für Train/Test
    params: Dict[str, Any] = LR_PRESET,     # Hyperparameter der LR
    with_standardization: Optional[bool] = None
) -> Dict[str, Any]:
    """
    Pipeline:
      1) lade/erzeuge deterministisches (X,y) mit N, d
      2) optional Standardisierung (empfohlen: withStd=True, withMean=False)
      3) trainiere LR (binomial, L-BFGS) und messe AUC (Train, optional Test)
    Notation:
      N = Anzahl Beobachtungen, d = Anzahl Features, X∈R^{N×d}, y∈{0,1}^N, w∈R^d, b∈R.
    """
    t0 = time.time()
    df = make_or_load_dataset(spark, N, d, seed=seed_data,
                              repartition=spark.sparkContext.defaultParallelism)

    # grobe Größenabschätzung (per JSON-Sample)
    def approx_df_bytes(df: DataFrame, max_rows: int = 1000, sample_fraction: float = 0.02) -> int:
        cnt = df.count()
        if cnt == 0: return 0
        frac = min(1.0, max(sample_fraction, min(1.0, max_rows / max(1, cnt))))
        local = df.sample(False, frac, seed=123).limit(max_rows).toJSON().collect()
        if not local: return 0
        avg = sum(len(x.encode("utf-8")) for x in local) / len(local)
        return int(avg * cnt)
    est_gb = approx_df_bytes(df) / (1024**3)

    # Standardisierung
    if with_standardization is None:
        with_standardization = bool(params.get("standardization", True))
    if with_standardization:
        scaler = StandardScaler(inputCol="features", outputCol="features_scaled", withStd=True, withMean=False)
        sc_model = scaler.fit(df)
        df = sc_model.transform(df).select(F.col("label"), F.col("features_scaled").alias("features"))

    # optional Split
    if test_fraction and 0.0 < test_fraction < 1.0:
        train_df, test_df = df.randomSplit([1.0 - test_fraction, test_fraction], seed=2024)
    else:
        train_df, test_df = df, None

    # LR fit
    lr = LogisticRegression(
        featuresCol="features", labelCol="label",
        regParam=float(params["regParam"]),
        elasticNetParam=float(params["elasticNetParam"]),
        maxIter=int(params["maxIter"]),
        tol=float(params["tol"]),
        standardization=bool(params["standardization"]),
        family=str(params["family"]),
        aggregationDepth=int(params["aggregationDepth"]),
        fitIntercept=bool(params["fitIntercept"])
    )

    t1 = time.time()
    model = lr.fit(train_df)
    fit_time = time.time() - t1

    # Metriken
    auc_train = evaluator_bin.evaluate(model.transform(train_df))
    auc_test = None
    if test_df is not None:
        auc_test = evaluator_bin.evaluate(model.transform(test_df))

    summary = getattr(model, "summary", None)
    total_iter = getattr(summary, "totalIterations", None)
    objective = getattr(summary, "objectiveHistory", None)

    return {
        "run_id": str(uuid.uuid4()),
        "timestamp": now_ts(),
        "mode": spark.sparkContext.master,
        "N": int(N),
        "d": int(d),
        "estimated_gb": float(est_gb),
        "fit_time_sec": float(fit_time),
        "total_time_sec": float(time.time() - t0),
        "metric": "AUC",
        "metric_value": float(auc_train),
        "metric_test_value": (float(auc_test) if auc_test is not None else None),
        "iterations": int(total_iter) if total_iter is not None else None,
        "regParam": float(params["regParam"]),
        "elasticNetParam": float(params["elasticNetParam"]),
        "maxIter": int(params["maxIter"]),
        "tol": float(params["tol"]),
        "standardization": bool(params["standardization"]),
        "family": str(params["family"]),
        "tag": None  # wird im Runner gesetzt
    }
```

```python
# [6] Vollständige Skalierung mit zwei Listen – geordnete Ausführung & eine CSV
from typing import Iterable, Tuple

def _to_list(x) -> List[int]:
    if x is None: return []
    if isinstance(x, (list, tuple, set, np.ndarray)): return [int(v) for v in x]
    return [int(x)]

def run_full_scaling(
    spark,
    *,
    label: str,                         # z. B. "1W", "2W"
    d_list: Iterable[int],
    N_list: Iterable[int],
    order: str = "N_then_d",            # oder "d_then_N"
    params: Dict[str, Any] = None,
    seed_data: int = 42,
    test_fraction: Optional[float] = None
) -> Tuple[pd.DataFrame, Path]:
    """
    Führt alle Kombinationen aus N_list × d_list aus – in definierter Reihenfolge.
    - order="N_then_d": äußere Schleife N, innere Schleife d  (empfohlen)
    - order="d_then_N": äußere Schleife d, innere Schleife N
    Persistiert genau EINE series.csv im Zeitstempel-Ordner.
    Notation: N = Anzahl Beobachtungen, d = Anzahl Features.
    """
    if params is None: params = LR_PRESET
    N_vals = _to_list(N_list)
    d_vals = _to_list(d_list)
    assert len(N_vals) > 0 and len(d_vals) > 0, "N_list und d_list dürfen nicht leer sein."
    assert order in ("N_then_d", "d_then_N"), "order ∈ {'N_then_d','d_then_N'}"

    out_dir = make_run_dir(label)
    csv_path = out_dir / "series.csv"
    rows: List[Dict[str, Any]] = []
    seq = 0

    if order == "N_then_d":
        outer_name, inner_name = "N", "d"
        for Ni in N_vals:
            for dj in d_vals:
                r = run_logreg_experiment(
                    spark, N=int(Ni), d=int(dj),
                    seed_data=seed_data, test_fraction=test_fraction, params=params
                )
                r.update({
                    "tag": label,
                    "order": order,
                    "outer": outer_name, "outer_value": int(Ni),
                    "inner": inner_name, "inner_value": int(dj),
                    "seq_idx": seq
                })
                seq += 1
                rows.append(r)
                print(f"Done: N={r['N']:,}  d={r['d']:<3}  ~{r['estimated_gb']:.3f} GB  "
                      f"-> fit={human(r['fit_time_sec'])} s  AUC(train)={human(r['metric_value'])}"
                      + (f"  AUC(test)={human(r['metric_test_value'])}" if r.get('metric_test_value') is not None else ""))
    else:
        outer_name, inner_name = "d", "N"
        for dj in d_vals:
            for Ni in N_vals:
                r = run_logreg_experiment(
                    spark, N=int(Ni), d=int(dj),
                    seed_data=seed_data, test_fraction=test_fraction, params=params
                )
                r.update({
                    "tag": label,
                    "order": order,
                    "outer": outer_name, "outer_value": int(dj),
                    "inner": inner_name, "inner_value": int(Ni),
                    "seq_idx": seq
                })
                seq += 1
                rows.append(r)
                print(f"Done: N={r['N']:,}  d={r['d']:<3}  ~{r['estimated_gb']:.3f} GB  "
                      f"-> fit={human(r['fit_time_sec'])} s  AUC(train)={human(r['metric_value'])}"
                      + (f"  AUC(test)={human(r['metric_test_value'])}" if r.get('metric_test_value') is not None else ""))

    df = pd.DataFrame(rows)
    df.to_csv(csv_path, index=False)
    print(f"Serie gespeichert: {csv_path}")
    return df, out_dir
```

```python
# [7] Aggregation & Laden
from glob import glob

def load_latest_series(label:str) -> Optional[pd.DataFrame]:
    paths = sorted(glob(str(RESULTS_DIR / f"*_{label}/series.csv")))
    if not paths: return None
    df = pd.read_csv(paths[-1])
    # Spalten-Normalisierung (ältere CSVs unterstützen)
    rename = {}
    if "n_rows" in df.columns: rename["n_rows"] = "N"
    if "n_features" in df.columns: rename["n_features"] = "d"
    if rename:
        df = df.rename(columns=rename)
    return df

def aggregate_labels(labels: List[str]) -> pd.DataFrame:
    parts = []
    for lab in labels:
        df = load_latest_series(lab)
        if df is not None:
            if "tag" not in df.columns:
                df["tag"] = lab
            parts.append(df)
    return pd.concat(parts, ignore_index=True) if parts else pd.DataFrame()

print("Aggregator: load_latest_series(label), aggregate_labels([labels])")
```

```python
# [8] Plot-Funktionen – optionaler sweep-Filter: "d" oder "N"

# [P1] Plot-Helfer – Export & Dateinamen

from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as _dt

# Robuste Verzeichnisse
PLOTS_DIR = (RESULTS_DIR / "plots") if 'RESULTS_DIR' in globals() else Path("./results/plots")
PLOTS_DIR.mkdir(parents=True, exist_ok=True)

def _ts_slug():
    try:
        return ts_slug()  # falls bereits definiert
    except NameError:
        return _dt.datetime.now().strftime("%Y%m%d_%H%M%S")

def _agg(df: pd.DataFrame, by, y: str, how: str="median") -> pd.DataFrame:
    how = how.lower()
    if how not in {"median","mean","min","max"}:
        how = "median"
    out = (df.groupby(by, as_index=False)[y].agg(how)
             .sort_values(by))
    return out

def _fmt_vals(vals, key):
    if vals is None:
        return f"{key}NA"
    if isinstance(vals, (list, tuple, set)):
        vals = list(vals)
        return f"{key}{vals[0]}" if len(vals)==1 else f"{key}multi"
    return f"{key}{vals}"

def _mk_fname(scale_axis: str, N_values, d_values, y: str):
    ts = _ts_slug()
    return f"{scale_axis}_ {_fmt_vals(N_values,'N')}_{_fmt_vals(d_values,'d')}_{y.replace('_sec','-s')}_{ts}.png".replace(' ','').lower()

def _annotate(fig, text: str):
    fig.text(0.01, 0.01, text, ha="left", va="bottom", fontsize=9)

def _maybe_export(fig, scale_axis: str, N_values, d_values, y: str, export: bool):
    if not export:
        return None
    fname = _mk_fname(scale_axis, N_values, d_values, y)
    out = PLOTS_DIR / fname
    fig.savefig(out, dpi=150, bbox_inches="tight")
    print(f"Gespeichert: {out}")
    return out

# [P2] Laufzeit vs. N  (d fix oder mehrere d-Werte)
def plot_time_vs_N(
    df: pd.DataFrame,
    *,
    d_values=None,           # int oder Iterable[int]
    tag_filter=None,         # z.B. ["1W","W"]
    y: str = "fit_time_sec",
    agg: str = "median",
    logx: bool = False,
    logy: bool = False,
    title: str = None,
    export: bool = False
):
    if d_values is not None and not isinstance(d_values, (list, tuple, set)):
        d_values = [d_values]
    data = df.copy()
    if tag_filter is not None:
        data = data[data["tag"].isin(list(tag_filter))]
    if d_values is not None:
        data = data[data["d"].isin(list(d_values))]
    if data.empty:
        print("plot_time_vs_N: Kein passender Datensatz.")
        return

    fig, ax = plt.subplots()
    for (tag, d), sub in data.groupby(["tag", "d"]):
        cur = _agg(sub, by=["N"], y=y, how=agg)
        ax.plot(cur["N"], cur[y], marker="o", label=f"{tag} | d={int(d)}")
    ax.set_xlabel("N (Anzahl Beobachtungen)")
    ax.set_ylabel("Zeit [s]" if y.endswith("_sec") else y)
    ax.set_title(title or f"Laufzeit vs. N (Agg: {agg})")
    ax.grid(True, which="both")
    if logx: ax.set_xscale("log")
    if logy: ax.set_yscale("log")
    ax.legend(ncol=2)
    plt.tight_layout()

    #desc = f"Skalierung: N | d={list(d_values) if d_values is not None else 'alle'} | Tags={list(tag_filter) if tag_filter else 'alle'} | Agg={agg}"
    #_annotate(fig, desc)
    _maybe_export(fig, "n-scale", N_values="multi", d_values=d_values, y=y, export=export)
    plt.show()

# [P3] Laufzeit vs. d  (N fix oder mehrere N-Werte)
def plot_time_vs_d(
    df: pd.DataFrame,
    *,
    N_values=None,           # int oder Iterable[int]
    tag_filter=None,         # z.B. ["1W","W"]
    y: str = "fit_time_sec",
    agg: str = "median",
    logx: bool = False,
    logy: bool = False,
    title: str = None,
    export: bool = False
):
    if N_values is not None and not isinstance(N_values, (list, tuple, set)):
        N_values = [N_values]
    data = df.copy()
    if tag_filter is not None:
        data = data[data["tag"].isin(list(tag_filter))]
    if N_values is not None:
        data = data[data["N"].isin(list(N_values))]
    if data.empty:
        print("plot_time_vs_d: Kein passender Datensatz.")
        return

    fig, ax = plt.subplots()
    for (tag, N), sub in data.groupby(["tag", "N"]):
        cur = _agg(sub, by=["d"], y=y, how=agg)
        ax.plot(cur["d"], cur[y], marker="o", label=f"{tag} | N={int(N):,}")
    ax.set_xlabel("d (Anzahl Features)")
    ax.set_ylabel("Zeit [s]" if y.endswith("_sec") else y)
    ax.set_title(title or f"Laufzeit vs. d (Agg: {agg})")
    ax.grid(True, which="both")
    if logx: ax.set_xscale("log")
    if logy: ax.set_yscale("log")
    ax.legend(ncol=2)
    plt.tight_layout()

    #desc = f"Skalierung: d | N={list(N_values) if N_values is not None else 'alle'} | Tags={list(tag_filter) if tag_filter else 'alle'} | Agg={agg}"
    #_annotate(fig, desc)
    _maybe_export(fig, "d-scale", N_values=N_values, d_values="multi", y=y, export=export)
    plt.show()

```
